# Import All Requirements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import os
```

find the text document name

```python
def text_document_name(PATH):
    notspamfiles = []
    spamfiles = []
    
    for textfile in tqdm(os.listdir(PATH)):
            if textfile[:3] =='spm':
                spamfiles.append(os.path.join(PATH,textfile ))
            else:
                notspamfiles.append(os.path.join(PATH,textfile))
                
            
    return spamfiles,notspamfiles
```

```python
#path of the training data text file folder 
path = "/home/sysadm/Documents/nlp_assignment/dataset/training-data" 
```

```python
#get the spam data set file name and  not spam data set file name of training data
text_files = text_document_name(path)
text_files 
```

Divided the text file into spam text file and not spam text file

```python
spam_file_location = text_files[0]
spam_file_location
```

```python
not_spam_file_location = text_files[1]
not_spam_file_location
```

Open the file in read mode

```python
# Open the file in read mode
def read_text(file):
    with open(file, 'r') as file:
        content = file.read()
    return content
```

```python
#one such exanple of this function 
read_text(not_spam_file_location[10])
```

Create data frame with two columns , one column for description in text file and another in spam or not

```python
def make_data_frame(self):
    content = []
    for i in range(len(self)):
        re_text = read_text(self[i])
        content.append(re_text)
    if self == spam_file_location:
        dict = {"Description":content,"Spam":["yes"] * len(self)}
        df = pd.DataFrame.from_dict(dict)
    else :
        dict = {"Description":content,"Spam":["no"] * len(self)}
        df = pd.DataFrame.from_dict(dict)
    return df
```

```python
train_spam = make_data_frame(spam_file_location)
train_spam
#train_spam.to_csv('./train_spam_data.csv', index=False)
```

```python
train_not_spam = make_data_frame(not_spam_file_location)
train_not_spam 
#train_not_spam .to_csv('./train_not_spam_data.csv', index=False)
```

concate two data frame

```python
import pandas as pd

# Concatenate the dataframes vertically
train_df = pd.concat([train_spam, train_not_spam], ignore_index=True)
train_df
```

```python
train_df['Spam'].value_counts()
```

count plot of the target variablre

```python
# Visualize the target variable over train data
ax = sns.countplot(x='Spam', data=train_df,palette='cool')
# Display count on top of each bar
for p in ax.patches:
    count = p.get_height()
    ax.annotate(f'{int(count)}', (p.get_x() + p.get_width() / 2.0, count), ha='center', va='bottom', fontsize=8) 
    plt.title("spam vs not spam ")
plt.show()
```
